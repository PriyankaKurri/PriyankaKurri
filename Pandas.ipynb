Merge Operations

```python
import pandas as pd
player= ['player1', 'player2', 'player3']
point = [12, 15, 20]
game = ['game1', 'game2', 'game3']
df1 = pd.DataFrame({'Players': player,'Points':point, 'Game':game })
df1
```

```python
import pandas as pd
player= ['player1', 'player5', 'player6']
Wicket = [2, 5, 4]
game = ['game1','game5', 'game6']
df2 = pd.DataFrame({'Players': player,'Wickets':Wicket, 'Game':game })
df2
```

```python
#Inner Merge
df1.merge(df2)

#Or

df1.merge(df2, on='Players' , how ='inner')
```

```python
#Left Merge
df1.merge(df2, on='Game' , how ='left')
```

```python
#Right merge
df1.merge(df2, on='Game' , how ='right')
```

```python
#Outer Merge
df1.merge(df2, on='Players' , how ='outer')
```

JOIN Operations

```python
import pandas as pd
player= ['player1', 'player2', 'player3']
point = [12, 15, 20]
game = ['game1', 'game2', 'game3']
df3 = pd.DataFrame({'Players': player,'Points':point, 'Games':game }, index= ['L2', 'L3', 'L4'])
df3
```

```python
import pandas as pd
player= ['player1', 'player5', 'player6']
Wicket = [2, 5, 4]
game = ['game1','game3', 'game4']
df4 = pd.DataFrame({'Player': player,'Wickets':Wicket, 'Game':game }, index= ['L1', 'L3', 'L4'])
df4
```

```python
#Inner Join
df3.join(df4, how= 'inner')
```

```python
df3.join(df4, how= 'right')
```

```python
df3.join(df4, how= 'left')
```

```python
df3.join(df4, how= 'outer')
```

COncatenate DataFrame Operations

```python
pd.concat([df3,df4])
```
